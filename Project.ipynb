```python
#Μετατροπή της στήλης Results από sting σε float για να μπορεί να γίνει μέτρηση

def convert_time_to_seconds(value):
    val = str(value).strip()
    #Έλεγχος για ακυρώσεις η μη έγκυρους χρόνους
    if any(word in val.lower() for word in ['did' , 'dsq', 'none']):
      return None

    parts = val.split(':')
#Ορισμός συνάρτησης μετατροπής
    try:
      if len(parts) == 3:
        hours = float(parts[0])
        minutes = float(parts[1])
        seconds = float(parts[2])
        return  (hours *3600) + (minutes * 60) + seconds
      elif len(parts) == 2:
        return float(parts[0]) * 60 + float(parts[1])
      else:
        return float(parts[0])
    except (ValueError, TypeError):
      return None
#Εφαρμογη στο DataFrame
df['Results_Seconds'] = df['Results'].apply(convert_time_to_seconds)
```

```python
#Εξέλιξη χροόνων στα 100M Freestyle
import matplotlib.pyplot as plt
import numpy as np


race_data = df[(df['Distance (in meters)'] == '100m') &
               (df['Stroke'] == 'Freestyle') &
               (df['Gender'] == 'Men') &
               (df['Rank'] == 1)]

#Ταξινόμηση ανά έτος
race_data = race_data.sort_values('Year')

#Γράφημα
plt.figure(figsize=(10,6))
plt.plot(race_data['Year'], race_data['Results_Seconds'], marker= 'o', color = 'blue')

#Τίτλοι
plt.title('100m Freestyle')
plt.grid(True, linestyle="--", alpha=0.7)
plt.xlabel('Έτος')
plt.ylabel('Χρόνος')
```

```python
#Ποσοστό διαφοράς από το 1912 εως το 2020

time_1912 = race_data[race_data['Year']==1912]['Results_Seconds'].values[0]
time_2020 = race_data[race_data['Year']==2020]['Results_Seconds'].values[0]

#Υπολογισμός διαφοράς από τον τύπο

per_drop = ((time_1912-time_2020)/time_1912) * 100

print(f'Χρόνος το 1912: {time_1912} δευτερόλεπτα')
print(f'Χρόνος το 2020: {time_2020} δευτερόλεπτα')
print(f'Το ποσοστό είναι: {per_drop}%')
```

```python
#Linear Regression
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

#Δημιουργία 2D πίνακα
x = race_data[['Year']]
y = race_data['Results_Seconds']

#Δημιουργία και εκπαίδευση του μοντέλου
model = LinearRegression()
model.fit(x,y)

#Πρόβλεψη
y_pred = model.predict(x)

#Σχεδίαση των δεδομένων
plt.figure(figsize=(10,6))
plt.scatter(x, y, color = 'blue')
plt.plot(x , y_pred , color = 'red')


#Τίτλοι
plt.title('Πρόβλεψη Χρόνων (Linear Regression)')
plt.xlabel('Έτος')
plt.ylabel('Χρόνος')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

#Εμφάνιση ακρίβειας
r2_score = model.score(x, y)
print(f"Ακρίβεια Μοντέλου: {r2_score:.2f}")
```

```python
#Προσπάθεια μελλοντικής πρόβλεψεις χρόνων

future_year = np.array([[2024]])
predicted_time = model.predict(future_year)

print(f'O χρόνος θα είναι:{predicted_time} δευτερόλπετα')
```

```python
#Πολυωνυμική Παλινδρόμηση
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

x = race_data[['Year']]
y = race_data['Results_Seconds']

#Δημιουργία του μοντέλου
poly_features = PolynomialFeatures()
x_poly = poly_features.fit_transform(x)

#Εκπαίδευση του μοντέλου
poly_model = LinearRegression()
poly_model.fit(x_poly, y)

#Υπολογισμός πρόβλεψης
y_poly_pred = poly_model.predict(x_poly)

#Σχεδίαση
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='teal')
plt.plot(x, y_poly_pred, color='red', linewidth=2)

#Τίτλοι
plt.title('Πρόβλεψη Χρόνων (Polynomial Regression)')
plt.xlabel('Έτος')
plt.ylabel('Δευτερόλεπτα')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()



#Έλεγχος ακρίβειας
poly_r2_score = poly_model.score(x_poly, y)
print(f"Ακρίβεια Μοντέλου: {poly_r2_score}")
```

```python
#Μελλοντική πρόβλεψη

year_2024= np.array([[2024]])
year_2024_poly = poly_features.transform(year_2024)

# Πρόβλεψη
pred_2024_poly = poly_model.predict(year_2024_poly)
print(f"Η πρόβλεψη το 2024: {pred_2024_poly[0]} δευτερόλεπτα")
```

```python
#Random Forest Regressor είναι καλύτερο στην πρόβλεψη σε σχέση με το Classifier
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor

#Δεδομένα
x = race_data[['Year']]
y = race_data['Results_Seconds']


#Εκπαίδευση μοντέλου
rf_model = RandomForestRegressor(n_estimators = 100, random_state = 42)
rf_model.fit(x,y)


#Πρόβλεψη
y_pred_rf = rf_model.predict(x)

#Σχεδίαση
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='teal')
plt.plot(x, y_pred_rf, color='orange', linewidth=2)


#Τίτλοι
plt.title('Πρόβλεψη Χρόνων (RandomForestRegressor)')
plt.xlabel('Έτος')
plt.ylabel('Δευτερόλεπτα')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

#Ακρίβεια μοντέλου
rf_r2 = rf_model.score(x, y)
print(f"Ακρίβεια Μοντέλου : {rf_r2}")
```

```python
# Πρόβλεψη με Random Forest
predicted_rf_2024 = rf_model.predict([[2024]])[0]

print(f"Η πρόβλεψη το 2024: {predicted_rf_2024} δευτερόλεπτα")
```

```python
#Οι 3 προβλέψεις για τους χρόνους το 2024
print(f"Η πρόβλεψη το 2024: {predicted_rf_2024} δευτερόλεπτα (Random Forest)")
print('')
print(f"Η πρόβλεψη το 2024: {pred_2024_poly[0]} δευτερόλεπτα (Polynomial Regression)")
print('')
print(f'Η πρόβλεψη το 2024: {predicted_time[0]} δευτερόλπετα (Linear Regression)')
print('')
#Το παγκόσμιο ρεκόρ στα 100 μέτρα ελεύθερο ανδρών κατέχει ο Κινέζος κολυμβητής Pan Zhanle με χρόνο 46.40
print('Άρα πιο κοντά έπεσε το Polynomial Regression')
```
